```python
import pandas as pd
pd.set_option('display.max_rows', 600)
pd.set_option('display.max_columns', 600)
train = pd.read_csv(r'C:\Users\ADMIN\Desktop\Kalapa\train.csv')
train.head(5)
```

```python
def missingvalue_table(train):
    mis_val=train.isnull().sum()
    mis_val_percent=round(train.isnull().mean()*100,2)
    mv_table=pd.concat([mis_val,mis_val_percent],axis=1)
    mv_table.columns=['Missing values','% of missing values']
    print ("The dataframe has " + str(train.shape[1]) + " columns and " + str(train.shape[0]) + " Rows.\n"      
            "Below is the summary of the number and percentage of missing value by column")
    return mv_table
missingvalue_table(train)
```

```python
def label_table(train):
    lmissing = train['label'].isnull().sum()
    count = train['label'].value_counts()
    percent = train['label'].value_counts()/(len(train)-lmissing)*100
    l_table = pd.concat([count,percent],axis=1)
    l_table.columns=['Count','Percentage']
    return l_table
label_table(train)
```

```python
def province_table(train):
    pmissing = train['province'].isnull().sum()
    count = train['province'].value_counts()
    percent = train['province'].value_counts()/(len(train)-pmissing)*100
    pmissing
    p_table = pd.concat([count,percent],axis=1)
    p_table.columns=['Count','Percentage']
    print ("There are " + str(train["province"].nunique()) + " provinces in this dataset.")
    return p_table
province_table(train)
```

```python
def district_table(train):
    dmissing = train['district'].isnull().sum()
    count = train['district'].value_counts()
    percent = train['district'].value_counts()/(len(train)-dmissing)*100
    d_table = pd.concat([count,percent],axis=1)
    d_table.columns = ['Count', 'Percentage']
    print ("There are " + str(train["district"].nunique()) + " districts in this dataset. Below is the preview of 20 most popular districts.")
    return d_table.head(20)
district_table(train)
```

```python
def set_prop(train, x):
        missing = train[x].isnull().sum()
        count = train[x].value_counts()
        percent = train[x].value_counts()/(len(train)-missing)*100
        table = pd.concat([count,percent],axis=1)
        table.columns = ['Count', 'Percentage']
        print ("There are " + str(train[x].nunique()) + " variables in " + x)
        return table
def set_all():
    all = {}
    for x in range(1, 58):
        propertyName = 'FIELD_' + str(x)
        all[propertyName] = set_prop(train, propertyName)
    return all
set_all()
```
