```python
import pandas as pd
pd.set_option('display.max_rows', 600)
pd.set_option('display.max_columns', 600)
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\Kalapa\test.csv')
df.head(5)
```

```python
def missingvalue_table(df):
    mis_val=df.isnull().sum()
    mis_val_percent=round(df.isnull().mean()*100,2)
    mv_table=pd.concat([mis_val,mis_val_percent],axis=1)
    mv_table.columns=['Missing values','% of missing values']
    print ("The dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "Below is the summary of the number and percentage of missing value by column")
    return mv_table
missingvalue_table(df)
```

```python
def province_table(df):
    pmissing = df['province'].isnull().sum()
    count = df['province'].value_counts()
    percent = df['province'].value_counts()/(len(df)-pmissing)*100
    pmissing
    p_table = pd.concat([count,percent],axis=1)
    p_table.columns=['Count','Percentage']
    print ("There are " + str(df["province"].nunique()) + " provinces in this dataset.")
    return p_table
province_table(df)
```

```python
def district_table(df):
    dmissing = df['district'].isnull().sum()
    count = df['district'].value_counts()
    percent = df['district'].value_counts()/(len(df)-dmissing)*100
    d_table = pd.concat([count,percent],axis=1)
    d_table.columns = ['Count', 'Percentage']
    print ("There are " + str(df["district"].nunique()) + " districts in this dataset. Below is the preview of 20 most popular districts.")
    return d_table.head(20)
district_table(df)
```
